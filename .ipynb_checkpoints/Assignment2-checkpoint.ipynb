# Task 3: Data Modelling

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,0:19]
y = data['ca_cervix']
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=33)
```

```python
X_train.shape
```

```python
X_test.shape
```

## K-Folds Cross-Validation

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5)
```

```python
for train_index, test_index in kf.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
```

## Classification

```python
from sklearn.metrics import f1_score
```

### K Nearest Neighbour

#### Fitting Model

```python
def knn(k):
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(k)
    total_f1_train = 0
    total_f1_test = 0
    
    # feed each fold into the KNN classifier
    for i, (train_index, test_index) in enumerate(kf.split(data)):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        knn.fit(X_train, y_train)
        
        # compute f1-score for training set
        y_train_pred = knn.predict(X_train)
        f1_train = f1_score(y_train, y_train_pred, average='macro')
        total_f1_train = total_f1_train + f1_train
        
        # compute f1-score for test set
        y_test_pred = knn.predict(X_test)
        f1_test = f1_score(y_test, y_test_pred, average='macro')
        total_f1_test = total_f1_test + f1_test
        
    # compute the average f1-score across all folds and amount of overfitting
    avg_f1_train = (total_f1_train/5).round(4)
    avg_f1_test = (total_f1_test/5).round(4)
    overfit = avg_f1_train - avg_f1_test
    
    return avg_f1_train, avg_f1_test, overfit
```

#### Tuning Parameters

n_neighbours (k)

```python
# select different k values
k = []
for i in range(5, 37):
    k.append(i)

# compute results for different k values
f1_train = []
f1_test = []
overfit = []
for value in k:
    result = knn(value)
    f1_train.append(result[0])
    f1_test.append(result[1])
    overfit.append(result[2])

# show results in table
d = {'k':k, 'f1_train':f1_train, 'f1_test':f1_test, 'overfit':overfit}
knn_result = pd.DataFrame(data=d)
knn_result
```

### Decision Tree

#### Fitting Model

```python
def d_tree(criterion, max_depth, min_samples_split, min_samples_leaf, max_features, max_leaf_nodes):
    from sklearn.tree import DecisionTreeClassifier
    dtree = DecisionTreeClassifier(criterion=criterion, 
                                   max_depth=max_depth, 
                                   min_samples_split=min_samples_split, 
                                   min_samples_leaf=min_samples_leaf, 
                                   max_features=max_features, 
                                   max_leaf_nodes=max_leaf_nodes)
    total_f1_train = 0
    total_f1_test = 0
    
    # feed each fold into the decision tree
    for i, (train_index, test_index) in enumerate(kf.split(data)):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        dtree.fit(X_train, y_train)
        
        # compute f1-score for training set
        y_train_pred = dtree.predict(X_train)
        f1_train = f1_score(y_train, y_train_pred, average='macro')
        total_f1_train = total_f1_train + f1_train
        
        # compute f1-score for test set
        y_test_pred = dtree.predict(X_test)
        f1_test = f1_score(y_test, y_test_pred, average='macro')
        total_f1_test = total_f1_test + f1_test
        
    # compute the average f1-score across all folds
    avg_f1_train = (total_f1_train/5).round(4)
    avg_f1_test = (total_f1_test/5).round(4)
    overfit = avg_f1_train - avg_f1_test
    
    return avg_f1_train, avg_f1_test, overfit
```

#### Tuning Parameters

criterion

```python
# select different criteria
criteria = ['gini', 'entropy']

# compute results for different criteria
f1_train = []
f1_test = []
overfit = []
for value in criteria:
    result = d_tree(value, None, 2, 1, None, None)
    f1_train.append(result[0])
    f1_test.append(result[1])
    overfit.append(result[2])

# show results in table
d = {'criterion':criteria, 'f1_train':f1_train, 'f1_test':f1_test, 'overfit':overfit}
dtree_result = pd.DataFrame(data=d)
dtree_result
```

max_depth

```python
# select different max_depth
max_depth = [None]
for i in range(2, 20):
    max_depth.append(i)

# compute results for different max_depth
f1_train = []
f1_test = []
overfit = []
for value in max_depth:
    result = d_tree("gini", value, 2, 1, None, None)
    f1_train.append(result[0])
    f1_test.append(result[1])
    overfit.append(result[2])

# show results in table
d = {'max_depth':max_depth, 'f1_train':f1_train, 'f1_test':f1_test, 'overfit':overfit}
dtree_result = pd.DataFrame(data=d)
dtree_result
```

### Logistic Regression

```python
# Creating a function for the models
def models(X_train, y_train):
    
    # Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0, max_iter=1000).fit(X_train, y_train)
    log_pred = log.predict(X_train)
    
    # Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    dtree = DecisionTreeClassifier(criterion='gini', random_state=0).fit(X_train, y_train)
    dtree_pred = dtree.predict(X_train)
    
    # Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion='entropy', random_state=0).fit(X_train, y_train)
    forest_pred = forest.predict(X_train)
    
    # Calculating the F1 score on the training data
    log_f1 = f1_score(y_train, log_pred, average='macro')
    dtree_f1 = f1_score(y_train, dtree_pred, average='macro')
    forest_f1 = f1_score(y_train, forest_pred, average='macro')
    
    # Printing the scores on the training data
    print("Train F1-Score on Logistic Regression is: {:.3f}".format(log_f1))
    print("Train F1-Score on Decision Tree is: {:.3f}".format(dtree_f1))
    print("Train F1-Score on Random Forest is: {:.3f}".format(forest_f1))
    
    return log, dtree, forest
```

```python
# Getting all of the models
model = models(X_train, y_train)
```

```python
#Classification report on the test data
from sklearn.metrics import classification_report
print("Results on Logistic Regression")
print(classification_report(y_test, model[0].predict(X_test)))
print("_______________________________________________________")

print("Results on Decision Tree")
print(classification_report(y_test, model[1].predict(X_test)))
print("_______________________________________________________")

print("Results on Random Forest")
print(classification_report(y_test, model[2].predict(X_test)))
```
