# Task 3: Data Modelling

```python
# Seperating the target and rest of the attributes
X = data.iloc[:,0:19]
y = data['ca_cervix']
print(X.shape)
print(y.shape)
```

## 3.1 K Nearest Neighbours

**Feature selection by Hill Climbing:**

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

col_num = X.shape[1]

new_Ind = []
cur_MaxScore = 0.0
col_Ind_Random = shuffle(range(0,col_num), random_state=9)

for cur_f in range(0,col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData = X.iloc[:, new_Ind].copy()
    X_train, X_test, y_train, y_test = train_test_split(newData, y, test_size=0.3, random_state=9)
    clf = KNeighborsClassifier(5)
    fit = clf.fit(X_train, y_train)
    cur_Score = clf.score(X_test, y_test)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
        print("Score with " + str(len(new_Ind)) + " selected features: " + str(cur_Score))
```

**Running a baseline algorithm first without parameter tuning:**

```python
# Importing necessary packages
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
```

```python
# Fitting the model
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)
train_pred = knn_clf.predict(X_train)
test_pred = knn_clf.predict(X_test)
train_accuracy = accuracy_score(y_train, train_pred)
test_accuracy = accuracy_score(y_test, test_pred)
print("Training accuracy: {:.3f}".format(train_accuracy))
print("Test accuracy: {:.3f}".format(test_accuracy))
print(plot_confusion_matrix(knn_clf, X_test, y_test))
```

As we can see there is underfitting since the train score is lower than test score. Our next step is to tune the parameters of K nearest neighbours by using the validation set.

**Tuning the parameters:**

```python
k = []
for i in range(3,12):
    k.append(i)
    
#Storing results for different k values
accuracy_train = []
accuracy_val = []
accuracy_test = []

for neighbors in k:
    knn_clf = KNeighborsClassifier(neighbors).fit(X_train, y_train)
    train_pred = knn_clf.predict(X_train)
    val_pred = knn_clf.predict(X_val)
    test_pred = knn_clf.predict(X_test)
    accuracy_train.append(accuracy_score(y_train, train_pred))
    accuracy_val.append(accuracy_score(y_val, val_pred))
    accuracy_test.append(accuracy_score(y_test, test_pred))
    
#Plotting the train and val score for different distances
line1, = plt.plot(k, accuracy_train, 'b', label='Training')
line2, = plt.plot(k, accuracy_val, 'r', label='Validation')

plt.legend()
plt.ylabel('Accuracy')
plt.xlabel('Distance')
plt.show() 
```

Choose the approprite value from the plot above that can reduce the amount of overfitting. Usually the smaller the gap between the training and validation line, the corresponding value of distance should be used. In this case we will choose k as 4 or 5.

```python
# Running the selected parameter on the testing set.
knn_clf = KNeighborsClassifier(5)
knn_clf.fit(X_train, y_train)
train_pred = knn_clf.predict(X_train)
test_pred = knn_clf.predict(X_test)
train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)
print("Training accuracy: {:.3f}".format(train_acc))
print("Test accuracy: {:.3f}".format(test_acc))
print(classification_report(y_test,test_pred))
print(plot_confusion_matrix(knn_clf, X_test, y_test))
```

## 3.2 Decision Tree

**Feature selection by Hill Climbing:**

```python
from sklearn.tree import DecisionTreeClassifier

col_num = X.shape[1]

new_Ind = []
cur_MaxScore = 0.0
col_Ind_Random = shuffle(range(0,col_num), random_state=9)

for cur_f in range(0,col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData = X.iloc[:, new_Ind].copy()
    X_train, X_test, y_train, y_test = train_test_split(newData, y, test_size=0.3, random_state=10)
    clf = DecisionTreeClassifier()
    fit = clf.fit(X_train, y_train)
    cur_Score = clf.score(X_test, y_test)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
        print("Score with " + str(len(new_Ind)) + " selected features: " + str(cur_Score))
```

**Running a baseline algorithm first without parameter tuning:**

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
train_pred = dt_clf.predict(X_train)
test_pred = dt_clf.predict(X_test)
train_accuracy = accuracy_score(y_train, train_pred)
test_accuracy = accuracy_score(y_test, test_pred)
print("Training accuracy: {:.3f}".format(train_accuracy))
print("Testing accuracy: {:.3f}".format(test_accuracy))
print(plot_confusion_matrix(dt_clf, X_test, y_test))
```

From the above confusion matrix we can see that there is a overfitting on the training data. Now we will try to reduce the overfitting by trying different values of the parameters.

**Displaying the decision tree:**

```python
from sklearn import tree

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt_clf, 
                   feature_names=X_train.columns,  
                   class_names=['negative', 'positive'],
                   filled=True)
```

From the graph above, we can visualize the important features.

**Tuning the parameters on validation set:**

```python
from itertools import product
criterion = ['gini', 'entropy']
max_depth = [3,4,5,6,7]
max_features = ['sqrt','log2']

for c,ma,mf in list(product(criterion,max_depth,max_features)):
    dt_clf = DecisionTreeClassifier(criterion=c, max_depth=ma, max_features=mf)
    dt_clf.fit(X_train, y_train)
    train_pred = dt_clf.predict(X_train)
    val_pred = dt_clf.predict(X_val)
    train_acc = accuracy_score(y_train, train_pred)
    val_acc = accuracy_score(y_val, val_pred)
    print("Training Accuracy: {:.3f}, Validation Accuracy: {:.3f}, Overfitting: {:.3f}, with c: {}, ma: {}, ms: {}".format(train_acc,val_acc,train_acc-val_acc,c,ma,mf))
```

From the result of tuning, we found that the optimal set of paramters is the one with the least amount of overfitting.

```python
dt_clf = DecisionTreeClassifier(criterion='gini', max_depth=6, max_features='log2')
dt_clf.fit(X_train, y_train)
train_pred = dt_clf.predict(X_train)
test_pred = dt_clf.predict(X_test)
train_accuracy = accuracy_score(y_train, train_pred)
test_accuracy = accuracy_score(y_test, test_pred)
print("Training accuracy: {:.3f}".format(train_accuracy))
print("Testing accuracy: {:.3f}".format(test_accuracy))
print(plot_confusion_matrix(dt_clf, X_test, y_test))
```
